```python
#importing openCV library
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

```python
#reading the image
img = cv2.imread("dog.jpg")
cv2_imshow(img)
```

```python
#converting original image to a greyscale image
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
#inverting the greyscale image
invert = cv2.bitwise_not(gray_img)
cv2_imshow(invert)
```

```python
#blurring the inverted greyscale image
blur = cv2.GaussianBlur(invert,(21,21),0)
cv2_imshow(blur)
```

```python
#final pencil sketch image
def sketch(i,j):
  return cv2.divide(i,255-j,scale=256)

final = sketch(gray_img,blur)
cv2_imshow(final)
```
