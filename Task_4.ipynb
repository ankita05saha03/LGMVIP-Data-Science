```python
#importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#reading the dataset
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='latin-1', error_bad_lines=False)
df.head()
```

```python
#describing the dataset
df.describe()
```

```python
#summary of the data set
df.info()
```

```python
#datatype of each column
df.dtypes
```

```python
#columns present in the dataset
df.columns
```

```python
df.columns.values
```

```python
#extracting the necessary columns
df=df[['eventid', 'iyear', 'imonth', 'country', 'region','provstate','city','crit1', 'crit2', 'crit3','success', 'suicide', 'attacktype1','targtype1','natlty1','gname','guncertain1','claimed','weaptype1','nkill','nwound']]
df.head()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#combining two columns('nkill' and 'nwound' into a new column 'casualities')
# converting 'nkill' and 'nwound' columns to numeric type
df['nkill'] = pd.to_numeric(df['nkill'], errors='coerce')
df['nwound'] = pd.to_numeric(df['nwound'], errors='coerce')

# filling missing values with 0
df['nkill'] = df['nkill'].fillna(0)
df['nwound'] = df['nwound'].fillna(0)

df['casualties'] = df['nkill'] + df['nwound']

print(df.isnull().sum())
```

```python
df.describe()
```

```python
print(f"""The dataset covers {df['country'].nunique()} countries across {df['region'].nunique()} regions, spanning {df['claimed'].nunique()} years from {df['iyear'].min()} to {df['iyear'].max()}.
A total of {df.index.nunique()} terrorist attacks have been recorded, resulting in {df['casualties'].sum()} casualties, including {df['nkill'].sum()} fatalities and {df['nwound'].sum()} wounded.""")
```

```python
plt.subplots(figsize=(15, 6))
sns.countplot(x='iyear', data=df, palette='RdYlGn_r', edgecolor=sns.color_palette('dark', 10))
plt.xticks(rotation=90)
plt.title("Number of Terrorist Activities Each Year")
plt.show()
```

```python
yearc = df[['iyear', 'casualties']].groupby('iyear').sum()
yearc.plot(kind='bar', color='red', figsize=(15, 6))
plt.title("Casualties")
plt.xlabel('Years', fontsize=15)
plt.ylabel('No. of Casualties', fontsize=15)
plt.show()
```

```python
df['attacktype1'].value_counts().plot(kind='bar',figsize=(20,10),color='yellow')
plt.xticks(rotation=50)
plt.xlabel('Attack Type',fontsize=20)
plt.ylabel('Number of attacks')
plt.title('Number of attacks')
plt.show()
```

```python
plt.subplots(figsize=(20, 10))
sns.countplot(x='targtype1', data=df, order=df['targtype1'].value_counts().index, palette='gist_heat', edgecolor=sns.color_palette("crest"))
plt.xticks(rotation=90)
plt.xlabel('Attack type', fontsize=20)
plt.ylabel('Count')
plt.title('Type of attack')
plt.show()
```

```python
plt.subplots(figsize=(15,6))
country_attacks = df.gname.value_counts()[:15].reset_index()
country_attacks.columns = ['gname', 'Total Attacks']
sns.barplot(x = country_attacks.gname, y = country_attacks['Total Attacks'], palette='OrRd_r', edgecolor=sns.color_palette('dark', 10))
plt.xticks(rotation = 90)
plt.title("Number of Total Attacks in Each Country")
plt.show()
```
